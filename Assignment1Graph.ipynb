We are trying to create a simple network. To do this we will use python networkx library. This library helps us create and analyse networds. Please pip install the library if it was not done before. Lets start with importing the library. 

```python
import networkx as nx
```

Here we start with the node Diane and build edges. After we are done we can view our results

```python
G = nx.Graph()

G.add_node("Diane")

G.add_edge("Diane","Carlos")
G.add_edge("Diane","Ed")
G.add_edge("Diane","Andre")
G.add_edge("Diane","Beverly")
G.add_edge("Diane","Garth")
G.add_edge("Diane","Fernando")

G.add_edge("Carlos","Andre")
G.add_edge("Carlos","Fernando")
G.add_edge("Carlos","Diane")

G.add_edge("Andre","Carlos")
G.add_edge("Andre","Fernando")
G.add_edge("Andre","Beverly")
G.add_edge("Andre","Diane")

G.add_edge("Beverly","Andre")
G.add_edge("Beverly","Diane")
G.add_edge("Beverly","Garth")
G.add_edge("Beverly","Ed")

G.add_edge("Ed","Beverly")
G.add_edge("Ed","Diane")
G.add_edge("Ed","Garth")

G.add_edge("Garth","Fernando")
G.add_edge("Garth","Heather")
G.add_edge("Garth","Ed")
G.add_edge("Garth","Beverly")
G.add_edge("Garth","Diane")

G.add_edge("Fernando","Carlos")
G.add_edge("Fernando","Andre")
G.add_edge("Fernando","Diane")
G.add_edge("Fernando","Garth")
G.add_edge("Fernando","Heather")

G.add_edge("Heather","Fernando")
G.add_edge("Heather","Ike")
G.add_edge("Heather","Garth")

G.add_edge("Ike","Heather")
G.add_edge("Ike","Jane")

nx.draw(G, with_labels=True)
```

Lets sort on the dress and view the results

```python
sorted(nx.degree(G))
```
